**PULLING DATA FROM PUBLIC API's - GET REQUEST**

```python
base_url = "http://api.exchangeratesapi.io/v1/latest?access_key=36cd09ce58578da2a5e3019c728f13cc&format=1"
```

**EXTRACTING THE DATA ON THE CURRENCY EXCHANGE RATE**

```python
# By importing the requests package, we have access to its ''.get()' method
# The '.get()' method submits a GET request to the indicated URL, and returns the response from the server

import requests
```

```python
# We stored it in a variable because we need the response to this request for further processing.

response = requests.get(base_url)
```

**INVESTIGATING THE RESPONSE**

```python
# We use the '.ok' attribute to check if the response went through without any problem

response.ok
```

```python
# We can also access the status code directly

response.status_code
```

```python
# To access the data contained in the text

response.text
```

```python
# To return it in bytes form

response.content
```

**HANDLING THE JSON**
- The data is in JSON format, which consists of dictionaries and lists.

-  It would be useful if instead of a string, we represent it in native python objects.

```python
# The 'requests' library provies us with the .json() method, which converts a JSON formatted response to a native Python object.

response.json()
```

```python
# Confirm the transformartion with the type function
# We can see that it has been converted to a python dictionary

type(response.json())
```

**JSON**
- json package provides methods for json manipulations

- 2 of the main ones are: Loads and dumps

- loads(string): converts a JSON formatted string to a Python object

- dumps(obj): converts a Python object back to a regular string, with options to make the string prettier

```python
import json
```

```python
#We can set an indentation for nested elements and choose the amount of white space with the indent parameter
# Here we used an indentation of 4

json.dumps(response.json(), indent = 4)
```

```python
# In order to manifest it, we need to print this string

print(json.dumps(response.json(), indent = 4))
```

```python
# We can take a look and inspoect the contents of this JSON
# It contains 4 keys as explained by the JSON method

response.json().keys()
```
